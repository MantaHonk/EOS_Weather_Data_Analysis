# Results

In a results section you present plots and tables that make some point. The supporting text should simply describe the plots, but interpretation is usually saved for the discussion section.

```python
# AI generated code for a little example

import numpy as np
import matplotlib.pyplot as plt

# Generate random data with a desired correlation
np.random.seed(42)  # Set seed for reproducibility
num_points = 30
x = np.random.rand(num_points)
y = 0.75 * x + np.random.rand(num_points) * 0.5  # Adjust 0.5 to control scatter

# Calculate correlation
correlation = np.corrcoef(x, y)[0, 1]

# Create scatterplot
plt.scatter(x, y)
plt.title(f"Figure 2: Scatterplot between Ottawa and Montréal")
plt.xlabel("X")
plt.ylabel("Y")

# Print correlation on the plot
plt.text(0.1, 0.9, f"Correlation: {correlation:.2f}", transform=plt.gca().transAxes)

#To see the output, run the code.
```

A strong positive relationship exists between the two variables, as indicated by a correlation of 0.75 (Fig. 2). This suggested that developing an empirical function whereby one could predict the other was reasonable.
